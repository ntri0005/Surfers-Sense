### Testing Google Geolocation API:

```python
%%time
GOOGLE_MAPS_API_KEY = ""

gmaps = googlemaps.Client(key = GOOGLE_MAPS_API_KEY)

# Geocoding an address
geocode_result = gmaps.geocode('Elliot Head, near Bundaberg. Queensland AUSTRALIA')
geocode_result
```

### Parsing all require attributes:

```python
lat = geocode_result[0]["geometry"]["location"]['lat']
lng = geocode_result[0]["geometry"]["location"]['lng']
location_type = geocode_result[0]["geometry"]["location_type"]
postal_code =  geocode_result[0]['address_components'][4]['long_name']
formatted_address = geocode_result[0]['formatted_address']
place_id = geocode_result[0]['place_id']
place_id
```

### Creating columns to save location: lat and long, location_type, postal_code, formatted_address and place_id

```python
raw_aus_shark_df["lat"] = ""
raw_aus_shark_df["lng"] = ""
raw_aus_shark_df["location_type"] = ""
raw_aus_shark_df["postal_code"] = ""
raw_aus_shark_df["formatted_address"] = ""
raw_aus_shark_df["place_id"] = ""
```

```python
%%time
gmaps = googlemaps.Client(key = GOOGLE_MAPS_API_KEY)

for index, row in raw_aus_shark_df.iterrows():
    address_str = row["Address"]
    print("index = ", index)
    try:
        # Geocoding an address

        geocode_result = gmaps.geocode(address_str)
        
        raw_aus_shark_df.at[index, "lat"] = geocode_result[0]["geometry"]["location"]['lat']
        raw_aus_shark_df.at[index,"lng"] = geocode_result[0]["geometry"]["location"]['lng']
        raw_aus_shark_df.at[index,"location_type"] = geocode_result[0]["geometry"]["location_type"]
        raw_aus_shark_df.at[index,"postal_code"] =  geocode_result[0]['address_components'][4]['long_name']
        raw_aus_shark_df.at[index,"formatted_address"] = geocode_result[0]['formatted_address']
        raw_aus_shark_df.at[index,"place_id"] = geocode_result[0]['place_id']
        
    except:
        
        print("address_str = ", address_str)
        
        pass    
```

### Pattern observed from above was that if I remove all tailing string after address from locaton then Google API will work:
* EXAMPLE: Stevens Reef, 70 miles from Mackay from that keep only "Stevens Reef"

```python
for index, row in raw_aus_shark_df[raw_aus_shark_df.lat == ""].iterrows():
    raw_aus_shark_df.at[index, 'Address'] = row['Location'].split(",")[0] +", "+ row['Area'] +", "+ row['Country']
raw_aus_shark_df[raw_aus_shark_df.lat == ""][['Location', 'Area', 'Country', 'Address']]
```

* Now address in better to be used by Google API
