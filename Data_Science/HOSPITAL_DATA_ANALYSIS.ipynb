```python
def convert_state_short_2_long(state):
    if state == "NSW":
        return "New South Wales"\
    
    elif state == "Vic":
        return "Victoria"
    
    elif state == "Qld":
        return "Queensland"
    
    elif state == "SA":
        return "South Australia"
    
    elif state == "WA":
        return "Western Australia"
    
    elif state == "Tas":
        return "Tasmania"
    
    elif state == "NT":
        return "Northern Territory"
    
    elif state == "ACT":
        return " Australian Capital Territory"
    
hos_wt_time_df["state"] = hos_wt_time_df["state"].apply(convert_state_short_2_long) 

hos_wt_time_df.state.unique()
```

```python
from math import radians, cos, sin, asin, sqrt

def calculate_distance_func(lat1, lon1, lat2, lon2):

    R = 6372.8 #  3959.87433 this is in miles.  For Earth radius in kilometers use 6372.8 km

    dLat = radians(lat2 - lat1)
    dLon = radians(lon2 - lon1)
    lat1 = radians(lat1)
    lat2 = radians(lat2)

    a = sin(dLat/2)**2 + cos(lat1)*cos(lat2)*sin(dLon/2)**2
    c = 2*asin(sqrt(a))

    return round(R * c, 3)
```

```python
common_states_lst = ['New South Wales', 
                     'Queensland', 
                     'Western Australia', 
                     'Tasmania',
                     'Victoria', 
                     'South Australia',
                    ]
```

```python
%%time
# Wall time: 1h 1min 5s
for state in common_states_lst:
    for index_o, row_o in beach_df.loc[beach_df["country_state"] == state, :].iterrows():
        for index_i, row_i in req_hos_wt_time_df.loc[req_hos_wt_time_df["state"] == state, :].iterrows():

            beach_hospital_row_dict = {}

            for beach_col in beach_col_lst:
                beach_hospital_row_dict[beach_col] = row_o[beach_col]

            for hospital_col in hospital_col_lst:
                beach_hospital_row_dict[hospital_col] = row_i[hospital_col]
    #         DISTANCE COL ENTRY
            beach_hospital_row_dict["beach_hospital_distance_km"] = calculate_distance_func(
                                                                        lat1 = row_i["hospital_latitude"], 
                                                                        lon1 = row_i["hospital_longitude"],
                                                                        lat2 = row_o["beach_latitude"], 
                                                                        lon2 = row_o["beach_longitude"])
            beach_hospital_df = beach_hospital_df.append(beach_hospital_row_dict, 
                                                         ignore_index = True)
        
```
