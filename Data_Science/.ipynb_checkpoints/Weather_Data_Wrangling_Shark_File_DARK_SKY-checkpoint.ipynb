# Clean Case_Date:

### Remove tailing letter from case dates:

```python
re.sub("(.[a-zA-Z])|(-[a-zA-Z])", '',    '2019.04.27-a') 
```

```python
def remove_tailing_letters_func(Case_Date):
    return re.sub("(.[a-zA-Z])|(-[a-zA-Z])", '',Case_Date)

raw_shark_df['Case_Date'] = raw_shark_df['Case_Date'].apply(remove_tailing_letters_func)
raw_shark_df.tail(25)
```

### Remove rows where month and day == "00":

```python
raw_shark_df.isnull().sum()
```

```python
%%time
for index, row in raw_shark_df.iterrows():
    month = row["Case_Date"].split(".")[1]
    day = row["Case_Date"].split(".")[2]
    
    if(month == "00") | (day == "00"):
        raw_shark_df.at[index, "Case_Date"] = None        
#         print(row["Case_Date"])
```

```python
%%time
# Wall time: 36min 47s --> 900 rows given 

API_KEY = ""
req_col_lst = ['Case_Date', 'lat', 'lng', "Address","date_req_format","Time",
               'Type',"Activity", "Name", "Gender", "Age", "Injury", "Fatal (Y/N)", 
                "Species ", "href", "postal_code", "place_id"]    
cnt = 0
# https://api.darksky.net/forecast/bdbb908566b674b2e5971ac7f61e1bfc/-33.869,151.209,2019-12-30T12:00:00
for index, row in raw_shark_df[req_col_lst].iterrows():
    if index <= 900:
        
        get_request = ""
        get_request += "https://api.darksky.net/forecast/"
        get_request += API_KEY
        get_request += "/"
        get_request += str(row["lat"])
        get_request += ","
        get_request += str(row["lng"])
        get_request += ","
        get_request += row['date_req_format']
        
        response_dict = json.loads(requests.get(get_request).text)
        
        print("get_request = ", get_request)
        print("****************************************************************************************************")
#         print("response_dict = ",response_dict)

        if len(list(response_dict.keys())) > 2:
            cnt = cnt + 1
            print("cnt =>>>>>>>>>>>>>>>>>>>>> ", cnt)
            weather_row_dict = {}
            try:
                for per_hr_attri_dict in response_dict['hourly']['data']:
                    for col in req_col_lst:
                        weather_row_dict[col] = row[col]

                    for hr_attri in per_hr_attri_dict.keys():
                        weather_row_dict[hr_attri] = per_hr_attri_dict[hr_attri]  

    #                 print("weather_row_dict = ", weather_row_dict)     

                    weather_row_dict["nearest-station"] = weat_dict['flags']['nearest-station']
                    weather_row_dict["sources"] = weat_dict['flags']['sources']
                    weather_row_dict["offset"] = weat_dict['offset']

                    weather_df = weather_df.append(weather_row_dict, ignore_index=True)
            except:
#                 print("response_dict = ", response_dict)
                print("Except case: ", row["lat"],row["lng"], row['date_req_format'])
                print("\n")
                pass
                                    
        print("\n")
        time.sleep(1)
# weather_df            
```

```python
agg_ops_lst = ['max','min','mean','median','std','var','sem']
numeric_features = ["temperature", "apparentTemperature", "dewPoint", "humidity", "windSpeed", "windBearing", "uvIndex",
                    "cloudCover"]

agg_dict = dict(zip(numeric_features, [agg_ops_lst for i in range(len(numeric_features))]))
agg_dict
```

```python
# agg_dict['summary'] = 'count'
```

```python
shark_weather_feat_df = weather_df.groupby('Date_Location').agg(agg_dict)

shark_weather_feat_df.columns = ["_".join(x) for x in shark_weather_feat_df.columns.ravel()]
# shark_weather_feat_df.columns = [col[1] for col in shark_weather_feat_df.columns]
# shark_weather_feat_df.columns = shark_weather_feat_df.columns.droplevel()
# shark_weather_feat_df.droplevel(1, axis = 1)
# shark_weather_feat_df.sum(level=1,axis=1)
# shark_weather_feat_df.columns = shark_weather_feat_df.columns.get_level_values(0)
shark_weather_feat_df.reset_index(level=0, inplace=True)
print("shark_weather_feat_df.shape = ", shark_weather_feat_df.shape)
shark_weather_feat_df
```

```python
shark_weather_feat_df.isnull().sum()
```

### Generating features from date:

```python
shark_weather_feat_df['month'] = shark_weather_feat_df['Date_Location'].apply(lambda x: x.split(".")[1])
shark_weather_feat_df['month_day'] = shark_weather_feat_df['Date_Location'].apply(lambda x: x.split(".")[2])
shark_weather_feat_df['month_day'] = shark_weather_feat_df['Date_Location'].apply(lambda x: 
                                                                                 pd.to_datetime(datetime(
                                                                                             int(x.split(".")[0]), 
                                                                                             int(x.split(".")[1]),
                                                                                             int(x.split(".")[2].split("_")[0])
                                                                                                         )
                                                                                                ).weekday()
                                                                                  )
def get_season_func(month):
    if (int(month) == 9) | (int(month) == 10) | (int(month) == 11):
        return "spring"
    elif (int(month) == 12) | (int(month) == 1) | (int(month) == 2):
        return "summer"
    elif (int(month) == 3) | (int(month) == 4) | (int(month) == 5):
        return "autumn"
    elif (int(month) == 6) | (int(month) == 7) | (int(month) == 8):
        return "winter"
    
shark_weather_feat_df['season'] = shark_weather_feat_df['month'].apply(get_season_func)
shark_weather_feat_df.head()
```
